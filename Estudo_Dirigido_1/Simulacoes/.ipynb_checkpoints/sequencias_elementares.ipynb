```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os

%matplotlib inline
```

Parametros

```python
N   = np.arange(-5, 21)         
n0  = 0                          
a   = 0.8                       
omega0 = np.pi / 4               
out_dir = 'resultados'
os.makedirs(out_dir, exist_ok=True)
```

```python
fig = plt.figure(figsize=(14, 12))
fig.suptitle('Sequências Elementares em Tempo Discreto', fontsize=15, fontweight='bold')

gs = gridspec.GridSpec(3, 2, hspace=0.55, wspace=0.35)
```

```python
#FUNCOES SEQUENCIAS
def impulso(n, n0=0):
    return (n == n0).astype(float)

def degrau(n, n0=0):
    return (n >= n0).astype(float)

def exp_real(n, a=0.8, n0=0):
    u = degrau(n, n0)
    return (a ** n) * u

def exp_complexa(n, omega0=np.pi/4):
    return np.exp(1j * omega0 * n)

def senoide(n, omega0=np.pi/4, A=1.0, phi=0.0):

    return A * np.cos(omega0 * n + phi)
```

```python
#CALCULO
delta  = impulso(N, n0)
u      = degrau(N, n0)
exp_r  = exp_real(N, a, n0)
exp_c  = exp_complexa(N, omega0)
sen    = senoide(N, omega0)
```

Plotagem

```python
fig.savefig(os.path.join(out_dir, 'sequencias_completas.png'),
            dpi=150, bbox_inches='tight')

plt.show()
```
