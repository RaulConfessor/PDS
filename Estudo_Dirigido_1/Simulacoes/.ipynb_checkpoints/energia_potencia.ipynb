```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os
```

```python
out_dir = 'resultados'
os.makedirs(out_dir, exist_ok=True)
```

```python
N_max = 200
n = np.arange(-N_max, N_max + 1)

def degrau(n): return (n >= 0).astype(float)
def impulso(n): return (n == 0).astype(float)
```

```python
#SINAIS
sinais = {
    r'$x_1[n]=(0.8)^n u[n]$':   (0.8 ** n) * degrau(n),
    r'$x_2[n]=u[n]$':           degrau(n),
    r'$x_3[n]=\cos(\pi n/4)$':  np.cos(np.pi * n / 4),
    r'$x_4[n]=n u[n]$':         n * degrau(n),
    r'$x_5[n]=\delta[n]$':      impulso(n),
}
```

```python
def energia(x):
    return np.sum(np.abs(x) ** 2)

def potencia(x):
    return np.sum(np.abs(x) ** 2) / len(x)

def classificar(E, P, tol=1e6):
    if E < tol and P < 1e-9:
        return 'ENERGIA'
    elif P > 1e-9 and P < tol:
        return 'POTÊNCIA'
    else:
        return 'NEM'
```

```python
resultados = {}
for nome, x in sinais.items():
    E = energia(x)
    P = potencia(x)
    classe = classificar(E, P)
    resultados[nome] = (x, E, P, classe)
```

```python
fig = plt.figure(figsize=(14, 14))
fig.suptitle('Energia e Potência de Sinais Discretos', fontsize=15, fontweight='bold')

gs = gridspec.GridSpec(4, 2, hspace=0.6, wspace=0.35)

n_zoom = np.arange(-15, 51)
```

```python
for i, (nome, (x_full, E, P, classe)) in enumerate(resultados.items()):
    ax = fig.add_subplot(gs[i // 2, i % 2])
    
    mask = (n >= n_zoom[0]) & (n <= n_zoom[-1])
    
    ml, sl, bl = ax.stem(n[mask], x_full[mask],
                         linefmt='C0-', markerfmt='C0o',
                         basefmt='k-')
    ml.set_markersize(4)

    ax.set_title(f'{nome}\n{classe} | E={E:.3g} | P={P:.4f}', fontsize=9)
    ax.set_xlabel('n', fontsize=8)
    ax.axhline(0, color='k', lw=0.5)
    ax.grid(True, alpha=0.3)
```

```python
n_pos = np.arange(0, 80)
x_exp_pos = (0.8 ** n_pos)
E_acum = np.cumsum(x_exp_pos ** 2)
E_analitica = 1 / (1 - 0.8**2)

ax_final = fig.add_subplot(gs[3, :])

ax_final.plot(n_pos, E_acum, 'b-o', markersize=4, label='Energia acumulada')
ax_final.axhline(E_analitica, color='r', ls='--',
                 label=f'Valor analítico = {E_analitica:.4f}')

ax_final.set_title('Convergência da Energia de x[n]=(0.8)^n u[n]', fontsize=11)
ax_final.set_xlabel('N')
ax_final.set_ylabel('Energia acumulada')
ax_final.legend()
ax_final.grid(True, alpha=0.3)
```

```python
plt.tight_layout()

fig.savefig(os.path.join(out_dir, 'sim3_completo.png'),
            dpi=150, bbox_inches='tight')

print("sim3_completo.png salvo em /resultados/")

plt.show()
```
