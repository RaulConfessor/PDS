```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os
```

```python
out_dir = 'resultados'
os.makedirs(out_dir, exist_ok=True)
```

```python
N = np.arange(0, 30)

def impulso(n): return (n == 0).astype(float)
def degrau(n):  return (n >= 0).astype(float)

x_imp = impulso(N)
x_deg = degrau(N)
```

```python
def S1(x, n):
    y = np.zeros(len(n))
    for i in range(len(n)):
        y[i] = x[i] + (x[i-1] if i > 0 else 0)
    return y

def S2(x, n):
    return x ** 2

def S3(x, n):
    y = np.zeros(len(n))
    for i in range(len(n) - 1):
        y[i] = x[i + 1]
    return y

def S4(x, n):
    return n * x

def S5(x, n):
    return np.cumsum(x)
```

```python
sistemas = [
    ('S1: $y[n]=x[n]+x[n-1]$', S1, 'Linear, IVT, Causal'),
    ('S2: $y[n]=x^2[n]$',      S2, 'Não-Linear, IVT, Causal'),
    ('S3: $y[n]=x[n+1]$',      S3, 'Linear, Não-Causal'),
    ('S4: $y[n]=n x[n]$',      S4, 'Variante no Tempo'),
    ('S5: $y[n]=\\sum x[k]$',  S5, 'Instável'),
]
```

```python
fig = plt.figure(figsize=(14, 16))
fig.suptitle('Classificação de Sistemas Discretos', fontsize=15, fontweight='bold')

gs = gridspec.GridSpec(5, 2, hspace=0.7, wspace=0.35)
```

```python
for i, (nome, sys, prop) in enumerate(sistemas):
    ax = fig.add_subplot(gs[i, 0])
    y = sys(x_imp, N)

    ml, sl, bl = ax.stem(N, y, basefmt='k-')
    ml.set_markersize(5)

    ax.set_title(f'{nome}\nImpulso → {prop}', fontsize=9)
    ax.set_xlabel('n')
    ax.grid(True, alpha=0.3)
```

```python
for i, (nome, sys, prop) in enumerate(sistemas):
    ax = fig.add_subplot(gs[i, 1])
    y = sys(x_deg, N)

    y_plot = np.clip(y, -5, 35)

    ml, sl, bl = ax.stem(N, y_plot, basefmt='k-')
    ml.set_markersize(5)

    nota = ' (clip)' if np.any(y != y_plot) else ''
    ax.set_title(f'{nome}\nDegrau{nota}', fontsize=9)
    ax.set_xlabel('n')
    ax.grid(True, alpha=0.3)
```

```python
plt.tight_layout()

fig.savefig(os.path.join(out_dir, 'sim4_completo.png'),
            dpi=150, bbox_inches='tight')

print("sim4_completo.png salvo em /resultados/")

plt.show()
```
