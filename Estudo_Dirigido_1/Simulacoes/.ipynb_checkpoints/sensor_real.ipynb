```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os
```

```python
out_dir = 'resultados'
os.makedirs(out_dir, exist_ok=True)

np.random.seed(2026)
```

```python
Fs = 1000
T = 1.0 / Fs
N_amst = 1000
n = np.arange(N_amst)
t = n * T
```

```python
f0 = 50.0
A1, A2 = 1.0, 0.4
sigma = 0.15

omega0 = 2 * np.pi * f0 / Fs

sinal_limpo = A1*np.cos(omega0*n) + A2*np.cos(2*omega0*n)
ruido = sigma * np.random.randn(N_amst)
sinal_sensor = sinal_limpo + ruido
```

```python
E = np.sum(sinal_sensor**2)
P = E / N_amst
```

```python
M = 5
def media_movel(x, M):
    h = np.ones(M)/M
    return np.convolve(x, h, mode='same')

sinal_filtrado = media_movel(sinal_sensor, M)
```

```python
X = np.fft.rfft(sinal_sensor)
freqs = np.fft.rfftfreq(N_amst, T)
mag = np.abs(X)/N_amst
```

```python
h = np.ones(M)/M
n_h = np.arange(len(h))

w = np.linspace(0, np.pi, 512)
H = np.array([np.sum(h * np.exp(-1j * wi * n_h)) for wi in w])
```

```python
fig = plt.figure(figsize=(14, 16))
fig.suptitle('Modelagem de Sinal de Sensor (PBL)', fontsize=15, fontweight='bold')

gs = gridspec.GridSpec(4, 2, hspace=0.6, wspace=0.35)

janela = slice(0, 200)
```

```python
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(n[janela], sinal_limpo[janela], 'b-')
ax1.set_title('Sinal Limpo')
ax1.grid(True, alpha=0.3)
```

```python
ax2 = fig.add_subplot(gs[1, :])
ax2.plot(n[janela], sinal_sensor[janela], 'r-', alpha=0.6, label='Com ruído')
ax2.plot(n[janela], sinal_filtrado[janela], 'b-', label='Filtrado')
ax2.legend()
ax2.set_title('Sinal com Ruído e Filtragem')
ax2.grid(True, alpha=0.3)
```

```python
ax3 = fig.add_subplot(gs[2, :])
ax3.plot(freqs, mag, 'g-')
ax3.set_xlim(0, 200)
ax3.axvline(f0, color='r', ls='--')
ax3.axvline(2*f0, color='b', ls='--')
ax3.set_title('Espectro de Frequência')
ax3.grid(True, alpha=0.3)
```

```python
ax4 = fig.add_subplot(gs[3, 0])
ml, sl, bl = ax4.stem(n_h, h)
ml.set_markersize(6)
ax4.set_title('Resposta ao Impulso do Filtro')
ax4.grid(True, alpha=0.3)
```

```python
ax5 = fig.add_subplot(gs[3, 1])
ax5.plot(w/np.pi, np.abs(H))
ax5.set_title('Resposta em Frequência')
ax5.grid(True, alpha=0.3)
```

```python
plt.tight_layout()

fig.savefig(os.path.join(out_dir, 'sim5_completo.png'),
            dpi=150, bbox_inches='tight')

print("sim5_completo.png salvo em /resultados/")
print(f"Energia: {E:.4f} | Potência: {P:.6f}")
```
